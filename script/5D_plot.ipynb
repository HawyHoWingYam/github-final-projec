```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

df = pd.read_csv('../data/train.csv')

# Create a new figure and a 3D subplot
fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111, projection='3d')

# Assign first three columns to axes
x = df['LotArea']
y = df['YearBuilt']
z = df['SalePrice']

# Use the fourth column for color
colors = df['TotalBsmtSF']  # Replace 'Column4' with the actual column name

# Use the fifth column for point size (normalize or scale as needed)
sizes = df['GarageArea']  # Replace 'Column5' with the actual column name
sizes = (sizes - sizes.min()) / (sizes.max() - sizes.min()) * 100  # Normalize sizes

# Create a scatter plot
sc = ax.scatter(x, y, z, c=colors, s=sizes, cmap='viridis', alpha=0.6, edgecolors='w')

# Add a color bar which maps values to colors
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('TotalBsmtSF')

# Label the axes
ax.set_xlabel('LotArea')
ax.set_ylabel('YearBuilt')
ax.set_zlabel('SalePrice')

# Show the plot
plt.show()
```
